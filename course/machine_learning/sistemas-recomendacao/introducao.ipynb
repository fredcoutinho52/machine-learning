```python
import numpy as np
import pandas as pd
```

```python
column_name = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_name)
```

```python
df.head()
```

```python
df.info()
```

```python
movie_title = pd.read_csv('Movie_Id_Titles')
```

```python
movie_title.head()
```

```python
df = pd.merge(df, movie_title, on='item_id')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
df_rating = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
df_rating.head()
```

```python
df_rating['count'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
df_rating.head()
```

```python
df_rating.shape
```

```python
plt.figure(figsize=(12, 8))
df_rating['count'].hist(bins=70)
```

```python
plt.figure(figsize=(12, 8))
df_rating['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating', y='count', data=df_rating, alpha=0.4, size=6)
```

```python
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat.head()
```

```python
moviemat.shape
```

```python
df_rating.sort_values('count', ascending=False).head()
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
```

```python
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
similar_starwars = moviemat.corrwith(starwars_user_ratings)
```

```python
similar_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_starwars, columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending=False).head()
```

```python
corr_starwars = corr_starwars.join(df_rating['count'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['count'] > 100].sort_values('Correlation', ascending=False).head()
```

```python
corr_liarliar = pd.DataFrame(similar_liarliar, columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(df_rating['count'])
```

```python
corr_liarliar[corr_liarliar['count'] > 100].sort_values('Correlation', ascending=False).head()
```
