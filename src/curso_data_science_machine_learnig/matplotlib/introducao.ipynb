```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# comando para mostrar o gráfico na mesma linha
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
x
```

```python
y = x * x
y
```

```python
# forma funcional
plt.plot(x, y, color="r")
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
```

```python
# subplot:
# 1 -> número de linhas
# 2 -> número de colunas
# 1 -> elemento que estou trabalhando
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure() # figura vazia

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.4])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Título')

axes2.plot(y, x)
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'r.-')

ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Título')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
# ax é um array
for axe in ax:
    axe.plot(x, y, 'y')
    axe.set_title('Título')

fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
fig
```

```python
fig, axes = plt.subplots(figsize=(12, 3))

axes.plot(x, y, 'r--', label='x ^ 2')
axes.plot(y, x, 'g--', label='x ** 1/2')
axes.set_title('Título')

axes.legend(loc=0) # loc=0 é o padrão
```

```python
fig.savefig("grafico.png") # salva uma imagem no formate png
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x, x**2, color="green", lw=5, alpha=0.5, linestyle=":")
ax.plot(x, x**3, color="blue", linewidth=0.5, marker="*")

ax.set_xlim([0, 2])
ax.set_ylim([0, 10])
```
