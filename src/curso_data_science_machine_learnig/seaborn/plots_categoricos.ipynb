```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', palette='rainbow', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, palette='rainbow', jitter=True, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='yellow')
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar')
```
