```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s': 100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=0.8, height=5)
```
