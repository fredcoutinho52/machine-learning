```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.info()
```

```python
graphics = sns.PairGrid(iris)
graphics.map(plt.scatter)
```

```python
graphics = sns.PairGrid(iris)
graphics.map_diag(plt.hist)
graphics.map_upper(plt.scatter)
graphics.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris, hue='species')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
graphics = sns.FacetGrid(tips, col='time', row='smoker')
graphics.map(plt.hist, 'total_bill')
```
