```python
import numpy as np

x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')

ax.plot(x, y)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.plot(x, y)
ax2.plot(x, y)

fig
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_xlim([0, 100])

ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20, 22])
ax2.set_ylim([30, 50])

ax1.plot(x, z)
ax2.plot(x, y)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, y, color="blue", linestyle="--")
ax[1].plot(x, z, color="red", lw="5")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))

ax[0].plot(x, y, color="blue", linestyle="--")
ax[1].plot(x, z, color="red", lw="5")
```
